```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/Users/amit/Documents/Data Science/Machine Lerning and Data science/EDA Pratice/Advance House Price Prediction/df_final.csv')
df.head()
```

```python
df.shape
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
X=df.drop(['Id','SalePrice'],axis=1)
X.head()
```

```python
y=df['SalePrice']
y.head()
```

For selelection of the best alpha value in the lasso model for feature selection 

```python
from sklearn.model_selection import cross_val_score
alpha=[i for i in np.arange(0.002,2,0.002)]

list_score=[]
for i in alpha:
    model_lasso=Lasso(alpha=i,random_state=2).fit(X,y)
    model_lasso_cv=cross_val_score(model_lasso,X,y,scoring='neg_mean_squared_error',cv=10)
    list_score.append([i , np.round(model_lasso_cv.mean(),4)])
```

```python
import operator
list_score.sort(key=operator.itemgetter(1))
```

```python
list_score
```

So we can assuming the alpha = 0.002 from the above accuracy list 

__Lasso Regression__ L1 regularaisation is used for the feature engineering part 

```python
selected_features=SelectFromModel(Lasso(alpha=0.002,random_state=2))
selected_features.fit(X,y)
```

```python
selected_features.get_support()      # True for selected and False for not selected
```

```python
selected_features=X.columns[(selected_features.get_support())]
```

```python
print(f'No of features in the original data : {X.shape[1]}')   # previous features 

print(f'No of selected features are : {len(selected_features)}')   
# features now selected after feature engineering
```

## Missing Values








__Note : Feature analysis already done in DataAnalysis[EDA] part__

__Categorical Features__

```python
feature_categorical_na=[feature for feature in df.columns if df[feature].dtypes=='O' and df[feature].isnull().sum()>=1]
print('Missing categorical features and their % of missing values')
for feature in feature_categorical_na:
    print(f'{feature} : {np.round(df[feature].isnull().mean(),4)}')
```

```python
for feature in feature_categorical_na:
    df[feature]=df[feature].fillna('Missing')  # will be taken care of later by encoding techniques 
```

```python
df[feature_categorical_na].isnull().sum()
```

__Numerical Features__

```python
features_numerical_na=[feature for feature in df.columns if df[feature].isnull().sum()>=1 and df[feature].dtypes!='O']
print('Missing numerical features and their % of missing values')
for feature in features_numerical_na:
    print(f'{feature} : {np.round(df[feature].isnull().mean(),4)}')
```

```python
for feature in features_numerical_na:
    mean_val=df[feature].median()        # replaced by median
    df[feature,'na']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(mean_val,inplace=True)
```

__3 New missing Numerical Features have been made by missing values as 1 and else 0__ 

__So as to caputure the more of the missing features from the numerical features__

```python
df[features_numerical_na].isnull().sum()
```

```python
df.shape
```

__Temporal Variables[DateTime variables]__

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]   # to find how much year from the year sold
```

```python
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt','YrSold']].head()
```

__Continous Variables [Conversion to log normal distribution]__

```python
features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in features:
    df[feature]=np.log(df[feature])
```

The continuous features that we saw in the EDA has now been converted into the log normal distribution 

So as to remove the skewness and outliers from the features data

## Handling Rare Categorical Features

```python
categorical_featues=[feature for feature in df.columns if df[feature].dtypes=='O']
for feature in categorical_featues:
    temp=df.groupby(feature)['SalePrice'].count()/len(df) 
    # taking the count of the features and then dividing by the total len of the data
                                                
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df) , df[feature] , 'RareVar') 
    # replacing unique values that are less then 0.01 from the categorical features
```

__The unique features of categorical variables occuring rarely is given the label 'RareVar'__

```python
print(len(categorical_featues))
```

## Encoding: 
As the categorical variable is Ordinal so using __Target Guided Ordinal Encoding__

The categorical variable having more mean wrt to the target variable [SalePrice] will be given higher label 

```python
label_list={}      # for later reference
for feature in categorical_featues:
    labels_ordered=df.groupby(feature)['SalePrice'].mean().sort_values().index  
    # unique categorical features are sorted wrt to mean of the target feature
    
    labels_ordered={k:i for i,k in enumerate(labels_ordered)}
    label_list[feature]=labels_ordered
    df[feature]=df[feature].map(labels_ordered)  # mapping of the categorical data to numerical 
```

## Feature Scaling

```python
# Features to be scaled
featues_to_scaled=[feature for feature in df.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[featues_to_scaled])
```

```python
df_final=pd.concat([df[['Id','SalePrice']] , pd.DataFrame(scaler.transform(df[featues_to_scaled]) , columns=featues_to_scaled)],axis=1)
```
